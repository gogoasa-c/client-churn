#### Outlier Detection

We will now calculate the interquartilic interval, and the first and third quartiles.

```python
q1 = data['Credit_Limit'].quantile(0.25)
q3 = data['Credit_Limit'].quantile(0.75)
iqr = q3 - q1
lower_lim = q1 - 3*iqr
upper_lim = q3 + 3*iqr
```

We will now extract the outliers, which are below the lower limit and above the upper limit.

```python
data[(data['Credit_Limit'] < lower_lim) | (data['Credit_Limit'] > upper_lim)]['Credit_Limit']
```

```python
fig, ax = plt.subplots(figsize = (8, 6))
sns.boxplot(x = data['Credit_Limit'], color = 'purple')
plt.show()
```

Checking for the outliers of each of the numerical columns.

```python
for col in num_columns:
    fig, ax = plt.subplots(figsize = (8, 6))
    sns.boxplot(x = data[col], color = 'purple')
    plt.show()
```

Given the above plots we can tell the following variables have outlier values:
* Credit_Limit
* Total_Unused_Bal
* Total_Amt_Chng_Q4_Q1
* Total_Trans_Amt
* Total_Ct_Chng_Q4_Q1


We will place these variables in a list:

```python
heavily_affected_by_outliers = ['Credit_Limit', 'Total_Unused_Bal', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Ct_Chng_Q4_Q1']
```

We will now be defining a function for censoring outlier values for a column of the DataFrame, so that the outlier values do not interfere with our analysis.

```python
def censoring_outliers(dataFrame, col):
    q1 = dataFrame[col].quantile(0.25)
    q3 = dataFrame[col].quantile(0.75)
    iqr = q3 - q1
    lower_lim = q1 - 3 * iqr
    upper_lim = q3 + 3 * iqr
    dataFrame[col] = np.where(
        dataFrame[col] < lower_lim,
        lower_lim,
        np.where(
            dataFrame[col] > upper_lim,
            upper_lim,
            dataFrame[col]
        )
    )
```

Now we'll be using the function defined above in order to actually censor all of those columns:

```python
for col in heavily_affected_by_outliers:
    censoring_outliers(data, col)
```
