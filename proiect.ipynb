We will now be defining a function for censoring outlier values for a column of the DataFrame, so that the outlier values do not interfere with our analysis.

```python
def censoring_outliers(dataFrame, col):
    q1 = dataFrame[col].quantile(0.25)
    q3 = dataFrame[col].quantile(0.75)
    iqr = q3 - q1
    lower_lim = q1 - 3 * iqr
    upper_lim = q3 + 3 * iqr
    dataFrame[col] = np.where(
        dataFrame[col] < lower_lim,
        lower_lim,
        np.where(
            dataFrame[col] > upper_lim,
            upper_lim,
            dataFrame[col]
        )
    )
```

Now we'll be using the function defined above in order to actually censor all of those columns:

```python
for col in heavily_affected_by_outliers:
    censoring_outliers(data, col)
```

### 4.2. Life and Gain Analysis

```python
# declare an empty DataFrame
lift_gain_report = pd.DataFrame()

# add y_test in DataFrame
lift_gain_report['y_test'] = y_test

# STEP 1
# add predict probabilities for class 1 in DataFrame
lift_gain_report['Predicted Probabilities'] = y_predict_proba_class_1

# order probabilities in increasing order
lift_gain_report['Probabilities Rank'] = lift_gain_report['Predicted Probabilities'].rank(method = 'first', ascending = True, pct = True)

# STEP 2
# calculate decile group
lift_gain_report['Decile Group'] = np.floor((1 - lift_gain_report['Probabilities Rank']) * 10) + 1

# STEP 3
# group observation by decile
lift_gain_report['Number of Observations'] = 1
lift_gain_report = lift_gain_report.groupby(['Decile Group']).sum().reset_index()

# cumulative number of observations
lift_gain_report['Cumulative no. of Observations'] = lift_gain_report['Number of Observations'].cumsum()

# cumulative percentage of observations
lift_gain_report['Cumulative % of no. of Observations'] = lift_gain_report['Cumulative no. of Observations']/lift_gain_report['Cumulative no. of Observations'].max()

# STEP 4
# calculate number of positives
lift_gain_report['Cumulative no. of Positives'] = lift_gain_report['y_test'].cumsum()

# STEP 5
# calculate cumulative percentage of positives - gain
lift_gain_report['Gain'] = lift_gain_report['Cumulative no. of Positives'] / lift_gain_report['Cumulative no. of Positives'].max()

# STEP 6
# calculate the lift
lift_gain_report['Lift'] = lift_gain_report['Gain'] / lift_gain_report['Cumulative % of no. of Observations']





lift_gain_report
```

By analyzing the table above we can tell that:
* by targeting the top 10% of the cases (those with the highest probability), we can reach 57.4% of the total positive values, meaning the churners. Thus, by targeting the first decile, we will be 5.72 times more efficient than by just selecting randomly.
* by targeting hte first 20% of the cases (those with the highest probability), we can reach 88.2% of the total positive values, meaning the churners. Thus, by targeting the first two deciles, we will be 4.4 times more efficient than by just selecting randomly.

The SHAP plot above shows us the impact that our variables have on whether or not a client will churn. These are ordered decreasingly by order of importance. Thus, the topmost vrariable is the one with most effect.

When the coloured dots take on values lesser than 0, the impact upon our model is a negative one, meaning it is less likely that our client will leave the bank.
