### Univariate Analysis

We want to analyze our target variable: Attrition_Flag


Checking for missing values:

```python
data['Attrition_Flag'].isnull().sum()
```

We can tell that there's no missing values for our target variables

```python
data['Attrition_Flag'].nunique() # checking the number of unique values
```

```python
data['Attrition_Flag'].unique() # checking the actual unique values
```

Viewing the frequency of each unique value as well as the percentage of the frequency represented

```python
data['Attrition_Flag'].value_counts()
```

```python
data['Attrition_Flag'].value_counts() / len(data['Attrition_Flag']) * 100
```

We can tell that out of the current customers:
* 83.93% of them are not attried (8500)
* 16.06% of them are attried (1627)

Visualizing the frequency distribution of the target variable:

```python
fig, ax = plt.subplots(figsize = (8, 6))
ax = sns.countplot(data = data, x = 'Attrition_Flag', palette = 'Set1')
plt.show()
```

We could now change the labels of the variable 'Attrition_Flag' to numeric values, for ease of use, using 0 for existing customer, 1 for attried customer.

```python
data['Attrition_Flag'] = data['Attrition_Flag'].map({'Existing Customer':0, 'Attrited Customer':1})
```

```python
data['Attrition_Flag'].value_counts()
```

### Feature Analysis

#### Categorical Variables

Exploring the categorical variables

```python
data.dtypes
```

```python
# selecting the categorical variables:
categ_columns = [col for col in data.columns if data[col].dtypes == 'object']
```

```python
print(f'There are {len(categ_columns)} categorical values. These are: {categ_columns}')
```

```python
data[categ_columns].head()
```

Checking the distribution of the categorical variables on their own

```python
for col in categ_columns:
    fix, ax = plt.subplots(figsize = (8, 6))
    ax = sns.countplot(data = data, x = col)
    plt.show()
```

#### Gender

For the Gender variable, we can tell that most of our clients are female.

#### Education Level

For the Education_Level variable, we can tell that most of our clients are graduates and the fewest of our clients are ones with a phD.

#### Marital Status

For the Mariat_Status variable, we can tell that most of our clients are married, closely followed by those that are single, and the fewest of all being those divorced.

#### Income Category

For the Income_Category variable, we can tell that most of our clients are in the less than $40K income bracket.

#### Card Category

For the Card_Category, we can tell that most of our clients have a blue card, and the fewest of our clients have a platinum card.

Checking the distribution of the categorical variables in relationship with the target variable 'Attrition_Flag'

```python
for col in categ_columns:
    fix, ax = plt.subplots(figsize = (8, 6))
    ax = sns.countplot(data = data, x = col, hue = 'Attrition_Flag')
    plt.show()
```

#### Missing Value Imputation

Checking which of the variables have missing values:

```python
data[categ_columns].isnull().sum()
```

Visualizing the columns:

```python
msno.bar(data[categ_columns])
```

```python
data['Education_Level'].value_counts()
```
